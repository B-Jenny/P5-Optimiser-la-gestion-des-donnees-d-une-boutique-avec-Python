```python
# d√©claration d'une fonction
def analyseDataFrame(df):
    """Fonction qui prend en arguments un dataframe pour analyse."""

    name =[x for x in globals() if globals()[x] is df][0]
    print("La dimension de DataFrame ¬´ " + name  +" ¬ª : " + str(df.shape) + " (lignes, colonnes)\n")
    display(temp_var = df.info())
    display(df.describe().T)
    display(df.head(5))
    
    # Afficher les colonnes qui sont unique dans le DataFrame df
    print("Test d'unicit√© sur les colonnes de DataFrame :")
    test_unicite = 0
    for col in df.columns :
        if df[col].size == df.drop_duplicates(col)[col].size :
          display(f'¬´ {col:s} ¬ª est unique, et pourrait √™tre une cl√© primaire.')
          test_unicite = 1
    if test_unicite == 0 :
        print("Il n'y a pas d'unicit√© dans aucune colonne. Il n'y a pas de cl√© primaire")

    print("Compter les nulles pour chaque colonne :")
    display(df.isnull().sum())
```

```python
# N'affichera plus le message de warning concernant excel
import warnings
warnings.filterwarnings('ignore', category=UserWarning, module='openpyxl')

# Importation des donn√©es au format excel
df_erp = pd.read_excel('erp.xlsx')
df_liaison = pd.read_excel('liaison.xlsx')
df_web = pd.read_excel('web.xlsx')
```

```python
# Cr√©ation d'un jeu de donn√©es sans les valeurs nulles de la colonne 'id_web'
df_liaison_notnull = df_liaison[df_liaison.id_web.notnull()]
analyseDataFrame(df_liaison_notnull)
```

### Donn√©es 'web'

```python
# Visualisation de web
analyseDataFrame(df_web)
```

```python
# Visualisation des donn√©es nulles de la colonne 'sku'
df_web[df_web.sku.isnull()]
```

```python
# les lignes 470 et 471 semblent pr√©senter une erreur √† v√©rifier avec l'√©quipe
df_web[df_web.sku.isnull() & df_web.post_name.notnull()]
```

```python
# Visualisation des duplicatas
df_web.sku.duplicated().sum()
```

```python
# Suppression des donn√©es dupliqu√©es de la colonne 'sku'
df_web = df_web.drop_duplicates(subset = ['sku'])
```

```python
# Cr√©ation d'un jeu de donn√©es sans les valeurs nulles de la colonne 'sku'
df_web_notnull = df_web[df_web.sku.notnull()]
analyseDataFrame(df_web_notnull)

# Cr√©ation d'un dataFrame contenant uniquement les valeurs 'product' de la colonne 'post_type'
df_produits = df_web_notnull[df_web_notnull['post_type']=='product']
analyseDataFrame(df_produits)
```

```python
# Cr√©ation d'une  jointure externe afin de permettre l'union des colonnes et le remplissage des valeurs nulles 
df_erp_liaison = pd.merge(df_erp, df_liaison_notnull,
                      on = 'product_id', how = 'outer', indicator = True)

# Visualisation du jeu de donn√©es erp_liaison
analyseDataFrame(df_erp_liaison)

# Visualisation du nombre de donn√©es unique de la colonne _merge
display(df_erp_liaison['_merge'].nunique())
```

#### La colonne '_merge' contient qu'une donn√©es unique : 'both'

```python
# Supression de la colonne '_merge'
df_erp_liaison = df_erp_liaison.drop(['_merge'], axis = 1)
```

```python
# Cr√©ation de la seconde jointure externe cette fois-ci avec des entr√©es de contenue √† droite et √† gauche
df_final = pd.merge(df_produits, df_erp_liaison, 
                      left_on = 'sku', right_on = 'id_web',  how = 'outer', indicator = True)


# Visualisation du jeu de donn√©es erp_lia_web
analyseDataFrame(df_final)

# Visualisation du nombre de donn√©es unique de la colonne _merge
display(df_final['_merge'].unique())
```

```python
# cr√©ation d'un jeu donn√©es avec uniquement les produits de donn√©es web
df_final_right = df_final[df_final['_merge'] == 'right_only']

# Visualisation du jeu de donn√©es erp_lia_web_right
analyseDataFrame(df_final_right)
```

```python
# Observation des lignes dont l'id_web n'est pas nul
df_id_web_notnull = df_final_right[df_final_right['id_web'].notnull()]

analyseDataFrame(df_id_web_notnull)
```

```python
# Supression de donn√©es 
df_final.drop(df_final[df_final['_merge'] != 'both'].index, inplace = True)

# Choix des colonnes
df_final = df_final[['sku', 'total_sales', 'product_id', 'id_web', 'price', 'post_title']]

# Visualisation de toutes les donn√©es de 'erp_lia_web'
analyseDataFrame(df_final)
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">Analyse univari√©e : M√©thode des interquartiles</h3>
</div>

```python
# Calcul interquartile
Q1= df_listing_vins['price'].quantile(q=0.25)
Q3= df_listing_vins['price'].quantile(q=0.75)  
IQR= Q3 - Q1 

# Calcul des valeurs extr√™mes
Lower = (Q1 - 1.5*IQR) 
Upper = (Q3 + 1.5*IQR)

# Outlier
outlier = df_listing_vins[(df_listing_vins['price'] > Upper) | (df_listing_vins['price'] < Lower)].sort_values('price')

# Visualisation
print('Q1                          :', Q1)
print('Q3                          :', Q3)
print('Borne minimum               :', Lower)
print('Borne maximum               :', Upper)
print('Etendue interquartile (IQR) :', IQR)
print("Nombre d'outliers           : " + str(len(outlier)))

# Visualisation des outliers
outlier
```

```python
sns.boxplot(x=outlier['price'], palette='pastel', )
```

```python
# Diagramme des prix en nuage des points (Les outliers en ROUGE)
df_produits_OK = df_listing_vins[(df_listing_vins['price'] < Upper) & (df_listing_vins['price'] > Lower)]
df_produits_outliers = df_listing_vins[(df_listing_vins['price'] > Upper) | (df_listing_vins['price'] < Lower)]

# On plot en rouge les produits outliers
plt.scatter(df_produits_OK['price'].index, df_produits_OK['price'].values)
plt.scatter(df_produits_outliers['price'].index, df_produits_outliers['price'].values,c='red')
plt.show()
```
